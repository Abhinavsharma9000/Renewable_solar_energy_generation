```python
import pandas as pd
import numpy as np


#change date here
date = '05102021'
input_path = 'C:/Users/eocn.re/Documents/aman-testing/day_ahead_forecast/forecasted_data/'
output_path  = 'C:/Users/eocn.re/Documents/aman-testing/day_ahead_forecast/R1_file/forecasted_GEN/'

dataframes= [pd.read_excel(p) for p in ((input_path+'ananthapur_gen_Forecast_'+date+'.xlsx'),
                                          (input_path+'rajgarh_gen_Forecast_'+date+'.xlsx'),
                                          (input_path+'bilhaur_gen_Forecast_'+date+'.xlsx'),
                                          (input_path+'bhadla_gen_Forecast_'+date+'.xlsx'),
                                          (input_path+'talcher_gen_Forecast_'+date+'.xlsx'),
                                          (input_path+'mandsaur_gen_Forecast_'+date+'.xlsx'),
                                          #(input_path+'unchahar_gen_Forecast_'+date+'.xlsx'),
                                          #(input_path+'rojmal_gen_Forecast_'+date+'.xlsx'),
                                          #(input_path+'noida_gen_Forecast_'+date+'.xlsx'),
                                          #(input_path+'kudgi_gen_Forecast_'+date+'.xlsx'),
                                        )]

dataframe = pd.concat(dataframes, axis=1)

# adding column name to the respective columns
dataframe.columns =['S2','ananthapur','S3','Rajgarh',
                    'S4','Bilhaur',
                    'S5','Bhadla',
                    'S6','Talcher','S7','mandsaur'
                    #'unchahar','S8','rojmal','S9','noida','S10','kudgi','S11'
                   ]
#drop unused columns
df = dataframe.drop(labels=['S2','S3',
                            'S4',
                            'S5','S6','S7'
                            #,'S8','S9','S10','S11'
                           ], axis = 1, inplace = False)

#SAVE all GEN file to one file
writer = pd.ExcelWriter(output_path+'All_GEN_'+date+'.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Generation')
writer.save()
```
