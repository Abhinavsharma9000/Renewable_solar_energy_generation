```python
import pandas as pd
import numpy as np


#change date here
date = '05102021'
input_path = 'C:/Users/eocn.re/Documents/aman-testing/day_ahead_forecast/daily_processed_ghi_data/'+date+'/'
test_path  = 'C:/Users/eocn.re/Documents/aman-testing/day_ahead_forecast/test_data/'

dataframes= [pd.read_csv(p) for p in ((input_path+'ananthapur_ghi_'+date+'.csv'),
                                      (input_path+'rajgarh_ghi_'+date+'.csv'),
                                      (input_path+'bilhaur_ghi_'+date+'.csv'),
                                      (input_path+'bhadla_ghi_'+date+'.csv'),
                                      (input_path+'talcher_ghi_'+date+'.csv'),
                                      (input_path+'mandsaur_ghi_'+date+'.csv'),
                                      (input_path+'unchahar_ghi_'+date+'.csv'),
                                      (input_path+'rojmal_ghi_'+date+'.csv'),
                                      (input_path+'noida_ghi_'+date+'.csv'),
                                      (input_path+'kudgi_ghi_'+date+'.csv'))]
dataframe = pd.concat(dataframes, axis=1)

# adding column name to the respective columns
dataframe.columns =['ananthapur', 'S2', 'Rajgarh','S3','Bilhaur','S4','Bhadla','S5','Talcher','S6','mandsaur','S7','unchahar',
                     'S8','rojmal','S9','noida','S10','kudgi','S11']
#drop unused columns
df = dataframe.drop(labels=['S2','S3','S4','S5','S6','S7','S8','S9','S10','S11'], axis = 1, inplace = False)

#SAVE all GHI file to one file
df.to_csv(test_path+'All_GHI_'+date+'.csv')
```
